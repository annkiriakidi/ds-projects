1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

x = symbols('x')
function = 10 * x - 5

function_derivative = diff(function, x)

f = lambda x: 10 * x - 5
f_prime = lambda x: 10

xs = np.linspace(0, 10, 100)
ys = f(xs)
ys_prime = np.full_like(xs, f_prime(0))

plt.figure(figsize=(10, 6))

plt.plot(xs, ys, label='Функція f(x) = 10x - 5', color='blue', linewidth=2)

plt.plot(xs, ys_prime, label="Похідна f'(x) = 10", color='orange', linestyle='--', linewidth=2)

plt.title('Графік функції та її похідної', fontsize=14)
plt.xlabel('Кількість годин навчання (x)', fontsize=12)
plt.ylabel('Оцінка та швидкість зміни оцінки', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.5)
plt.legend(fontsize=12)

plt.show()

print(f"Функція: {function}")
print(f"Похідна: {function_derivative}")
```

**Висновки**

Похідна дорівнює 10. Це означає, що швидкість росту оцінки постійна. Тобто скільки б не вчився -- швидкість зміни оцінки не буде збільшуватись.
Якщо кількість годин навчання збільшується на одну годину, оцінка завжди збільшується на 10 балів, без жодних змін у швидкості цього зростання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, sin, ln

x = symbols('x')

function_sin = sin(x)
function_ln = ln(x)

function_sin_derivative = diff(function_sin, x)
function_ln_derivative = diff(function_ln, x)

f_sin = lambda x: np.sin(x)
f_sin_prime = lambda x: np.cos(x)

f_ln = lambda x: np.log(x)
f_ln_prime = lambda x: 1 / x

xs = np.linspace(0.1, 10, 500)
ys_sin = f_sin(xs)
ys_sin_prime = f_sin_prime(xs)

ys_ln = f_ln(xs)
ys_ln_prime = f_ln_prime(xs)

plt.figure(figsize=(14, 10))

plt.subplot(2, 1, 1)
plt.plot(xs, ys_sin, label='Функція f(x) = sin(x)', color='blue', linewidth=2)
plt.plot(xs, ys_sin_prime, label="Похідна f'(x) = cos(x)", color='orange', linestyle='--', linewidth=2)
plt.title('Графік функції sin(x) та її похідної', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Значення', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.5)
plt.legend(fontsize=12)

plt.subplot(2, 1, 2)
plt.plot(xs, ys_ln, label='Функція f(x) = ln(x)', color='green', linewidth=2)
plt.plot(xs, ys_ln_prime, label="Похідна f'(x) = 1/x", color='red', linestyle='--', linewidth=2)
plt.title('Графік функції ln(x) та її похідної', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Значення', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.5)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

print(f"Функція: {function_sin}")
print(f"Похідна: {function_sin_derivative}")
print(f"Функція: {function_ln}")
print(f"Похідна: {function_ln_derivative}")
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

x = symbols('x')

function = x**2 - 5*x + 1

function_derivative = diff(function, x)

f = lambda x: x**2 - 5*x + 1
f_prime = lambda x: 2*x - 5

xs = np.linspace(0, 100, 500)
ys = f(xs)
ys_prime = f_prime(xs)

plt.figure(figsize=(10, 6))

plt.plot(xs, ys, label='Функція f(x) = x^2 - 5x + 1', color='blue', linewidth=2)

plt.plot(xs, ys_prime, label="Похідна f'(x) = 2x - 5", color='orange', linestyle='--', linewidth=2)

plt.title('Графік функції та її похідної', fontsize=14)
plt.xlabel('Кількість вироблених одиниць (x)', fontsize=12)
plt.ylabel('Значення', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.5)
plt.legend(fontsize=12)

plt.show()

print(f"Функція: {function}")
print(f"Похідна: {function_derivative}")
```
